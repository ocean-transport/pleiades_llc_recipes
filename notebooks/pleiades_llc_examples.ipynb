# Pleaides LLC Examples

Here are a few simple examples of using the `llcreader` module to work with LLC data on Pleaides.

Full documentation for `llcreader` functionality is online: <https://xmitgcm.readthedocs.io/en/latest/llcreader.html>

```python
import xmitgcm.llcreader as llcreader
%matplotlib inline
import holoviews as hv
from holoviews.operation.datashader import regrid
hv.extension('bokeh')
```

```python
model = llcreader.PleiadesLLC2160Model()
# to see all available variables
model.varnames
```

### Get a dataset

We use the `latlon` keyword to get rid of the Arctic and make a nice rectangular pseudo lat-lon grid.

```python
ds = model.get_dataset(varnames=['Eta'], type='latlon')
```

### Interactive Visualization

```python
dataset = hv.Dataset(ds.Eta.astype('f4'))
hv_im = (dataset.to(hv.Image, ['i', 'j'], dynamic=True)
                .options(cmap='RdBu_r', width=950, height=600, colorbar=True))

%output holomap='scrubber' fps=3
regrid(hv_im, precompute=True)
```

### Export a Dataset

Here is an example of how we might construct and export a reduced dataset.

```python
# select a region of interest
# it has to be done in i, j space, not lat and lon
region = ds.sel(i=slice(1000, 2000), j=slice(2500, 3500),
                i_g=slice(1000, 2000), j_g=slice(2500, 3500))

# convert to daily average
daily = region.resample(time='1D').mean()

# just the first 10 days

daily_10days = daily.isel(time=slice(0, 10))
```

```python
daily_10days.nbytes / 1e6
```

```python
from dask.diagnostics import ProgressBar

with ProgressBar():
    daily_10days.to_netcdf('/nobackup/rpaberna/llc_test_output.nc')
```

```python
# verify
import xarray as xr
xr.open_dataset('/nobackup/rpaberna/llc_test_output.nc')
```

### Other things you might want to do

Spatial sub-sampling to one degre

```python
ds_subsample = ds.isel(i=slice(None, None, 24), j=slice(None, None, 24))
ds_subsample
```

Spatial coarse grain to one degree

```python
eta_coarse = ds.Eta.coarsen(i=24, j=24).mean()
eta_coarse
```
